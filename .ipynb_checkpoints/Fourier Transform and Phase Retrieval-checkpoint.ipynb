```python
%pylab --no-import-all inline
from matplotlib.colors import LogNorm, SymLogNorm
from astropy.io import fits
matplotlib.rcParams['image.cmap'] = 'gray'
matplotlib.rcParams['image.origin'] = 'lower'
matplotlib.rcParams['image.interpolation'] = 'nearest'
matplotlib.rcParams['font.family'] = 'serif'
```

> # Starting from the JWST pupil, calculate its Fourier transform and plot the corresponding PSF.

OPD files and a FITS pupil image are available in `/grp/jwst/wit/miri/lajoie`.

```python
!cp /grp/jwst/wit/miri/lajoie/*.fits ./
!ls *.fits
```

```python
pupil = fits.getdata('./pupil.fits').astype(np.complex128, casting='safe')
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.real(pupil))
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

```python
def zoom_it(frac, imgdata):
    centerx, centery = imgdata.shape[:2]
    centerx /= 2
    centery /= 2
    frac = 1.0 / frac
    plt.xlim(centerx - (frac * float(centerx)), centerx + (frac * float(centerx)))
    plt.ylim(centery - (frac * float(centery)), centery + (frac * float(centery)))
    plt.scatter((centerx,), (centery,), marker='+', color='white', s=90, alpha=0.5)
```

```python
psf = np.fft.fftshift(np.fft.fft2(pupil))
```

```python
plt.figure(figsize=(10,10))
img = plt.imshow(np.abs(psf) / np.max(np.abs(psf)), norm=LogNorm(), cmap='gray_r')
cbar = plt.colorbar()
cbar.set_label('Relative Intensity')
zoom_it(6, psf)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

> # Why can't we simply use `numpy.fft.fft2`?

We aren't sampling the PSF very finely in the region we are most interested in: that of highest intensity in the core. If we have a pupil that spans the input array from edge to edge, we'll have a very small PSF. Instead, we must pad the pupil array in order to get more detail in our PSF output.

```python
# if we pad by 512, we preserve the useful property of dimensions being a power of two for the FFT
pad_pupil = np.pad(pupil, 512, mode='constant', constant_values=(0.,))
print pad_pupil.shape, pad_pupil.dtype
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.real(pad_pupil))
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

```python
pad_psf = np.fft.fftshift(np.fft.fft2(pad_pupil))
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1, 2, 1)
plt.title(u'PSF (pupil occupied 1024×1024 of 1024×1024 input array)')
img = plt.imshow(np.abs(psf) / np.max(np.abs(psf)), norm=LogNorm(), cmap='gray_r')
zoom_it(10, psf)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')

plt.subplot(1, 2, 2)
plt.title(u'PSF (pupil occupied 1024×1024 of 2048×2048 input array)')
img = plt.imshow(np.abs(pad_psf) / np.max(np.abs(pad_psf)), norm=LogNorm(), cmap='gray_r')
zoom_it(10, pad_psf)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

> #Implement the Gerchberg-Saxton algorithm to retrieve the phase across the pupil plane.

First, we use an OPD (optical path difference) map to create a PSF incorporating realistic phase information. The E field in the pupil is related to the pupil or aperture function and the OPD map as follows:

$$E_{ij} = \mathrm{Pupil}_{ij} \times \exp \left(i * 2 \pi * \frac{\mathrm{OPD}_{ij}}{\lambda}\right)$$

(For this exercise, $\lambda = 1$.)

```python
input_opd_map = fits.getdata('./MIRI_OPDisim6TiltCorrected.fits')
```

```python
pad_input_opd_map = np.pad(input_opd_map, 512, mode='constant', constant_values=(0.,))
plt.figure(figsize=(10, 10))
plt.imshow(np.ma.masked_array(pad_input_opd_map, mask=(pad_pupil <= 0)), cmap='rainbow')
cbar = plt.colorbar()
cbar.set_label('Optical Path Difference')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
plt.xlim(512,1536)
plt.ylim(512,1536)
```

```python
input_e_field = pad_pupil * np.exp(1.0j * 2 * np.pi * pad_input_opd_map)
print input_e_field.shape, input_e_field.dtype
```

```python
initial_psf = np.fft.fftshift(np.fft.fft2(input_e_field))
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1, 2, 1)
plt.title(u'PSF with OPD map (pupil occupied 1024×1024 of 2048×2048 input array)')
img1 = plt.imshow(np.real(initial_psf) / np.max(np.abs(initial_psf)), norm=LogNorm(), cmap='gray_r')
img1.set_clim(1e-9, 1.0)
zoom_it(10, initial_psf)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')

plt.subplot(1, 2, 2)
plt.title(u'PSF (pupil occupied 1024×1024 of 2048×2048 input array)')
img2 = plt.imshow(np.abs(pad_psf) / np.max(np.abs(pad_psf)), norm=LogNorm(), cmap='gray_r')
img2.set_clim(1e-9, 1.0)
zoom_it(10, pad_psf)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

Now we have a real-valued point-spread function (`initial_e_magnitude`), simulating one that we would see on a detector. The phase information is lost in the conversion to a real valued image, so we're left to make an initial guess at the phase in the intensity image.

We have $|\vec{E}|$, so to get $\vec{E}$ we need to compute $$\vec{E} = \mathrm{Aperture}_{ij} \ e^{i \phi}$$ where $\phi$ is a phase from $-\pi$ to $\pi$.

Below I do the algorithm 'by hand' (not in a loop) once to work out the exact steps required.

```python
# phase_guess = np.random.uniform(-1*np.pi, 1*np.pi, size=initial_psf.shape)
phase_guess = np.pi * np.ones(pad_pupil.shape)
print np.max(phase_guess), np.mean(phase_guess), np.min(phase_guess)

exp_phase_guess = np.exp(1.0j * phase_guess)
print np.max(pad_pupil), np.mean(pad_pupil), np.min(pad_pupil)

undo_exp = np.log(exp_phase_guess) / 1.0j
print np.max(undo_exp), np.mean(undo_exp), np.min(undo_exp), undo_exp.dtype

finfo = np.finfo(np.complex128)

back_out = np.log(np.real(pad_pupil) * exp_phase_guess) / 1.0j
back_out_ma = np.ma.masked_array(back_out, mask=(pad_pupil <= 0))
print np.max(back_out), np.mean(back_out), np.min(back_out)
print np.max(back_out_ma), np.mean(back_out_ma), np.min(back_out_ma)
print back_out_ma[back_out_ma == np.max(back_out_ma)]
print back_out[1100:1300,550:750]
```

```python
e_for_pupil = pad_pupil * np.exp(1.0j * phase_guess)
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(121)
plt.imshow(np.abs(e_for_pupil))
plt.xlim(512, 1536)
plt.ylim(512, 1536)
plt.colorbar()

plt.subplot(122)
real_part_e, imag_part_e = np.real(e_for_pupil), np.imag(e_for_pupil)
real_part_e[real_part_e == 0] = 1e-9
recovered_phase = np.arctan(imag_part_e / real_part_e)
plt.imshow(np.ma.masked_array(recovered_phase, mask=(pad_pupil <= 0)), cmap='jet')
print np.max(recovered_phase), np.mean(recovered_phase), np.min(recovered_phase)
plt.xlim(512, 1536)
plt.ylim(512, 1536)
plt.colorbar()

plt.figure(figsize=(10, 10))
plt.imshow(np.ma.masked_array(phase_guess, mask=(pad_pupil <= 0)), cmap='jet')
plt.xlim(512, 1536)
plt.ylim(512, 1536)
plt.colorbar()
```

```python
e_for_image_plane = np.fft.fftshift(np.fft.fft2(e_for_pupil))
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1, 2, 1)
plt.title(u'PSF from guessed phase (pupil occupied 1024×1024 of 2048×2048 input array)')
img2 = plt.imshow(np.abs(e_for_image_plane) / np.max(np.abs(e_for_image_plane)), cmap='gray_r')
# img2.set_clim(1e-9, 1.0)
plt.colorbar()
zoom_it(10, e_for_image_plane)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')

plt.subplot(1, 2, 2)
plt.title(u'PSF with OPD map (pupil occupied 1024×1024 of 2048×2048 input array)')
img1 = plt.imshow(np.abs(initial_psf) / np.max(np.abs(initial_psf)), norm=LogNorm(), cmap='gray_r')
img1.set_clim(1e-9, 1.0)
zoom_it(10, initial_psf)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```
