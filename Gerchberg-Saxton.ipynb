> #Implement the Gerchberg-Saxton algorithm to retrieve the phase across the pupil plane.

First, we use an OPD (optical path difference) map to create a PSF incorporating realistic phase information. The E field in the pupil is related to the pupil or aperture function and the OPD map as follows:

$$E_{ij} = \mathrm{Pupil}_{ij} \times \exp \left(i * 2 \pi * \frac{\mathrm{OPD}_{ij}}{\lambda}\right)$$

(For this exercise, $\lambda$ can be between 5.0 and 28.3 microns. The OPD is in units of microns, making the value in the exponential appropriately unitless.)

```python
pupil = fits.getdata('./pupil.fits')
```

```python
input_opd_map = fits.getdata('./MIRI_OPDisim6TiltCorrected.fits') / ((5.0 + 28.3) / 2.0) # MIRI goes from 5.0 to 28.3, so pick the middle
```

```python
pad_pupil = np.pad(pupil, 512, mode='constant', constant_values=(0.,))
pad_input_opd_map = np.pad(input_opd_map, 512, mode='constant', constant_values=(0.,))
```

```python
def zoom_it(frac, imgdata):
    """
    Zoom in to the center of the last-displayed image. (e.g. frac=2 is a 2x zoom)
    """
    centerx, centery = imgdata.shape[:2]
    centerx /= 2
    centery /= 2
    frac = 1.0 / frac
    plt.xlim(centerx - (frac * float(centerx)), centerx + (frac * float(centerx)))
    plt.ylim(centery - (frac * float(centery)), centery + (frac * float(centery)))
    plt.scatter((centerx,), (centery,), marker='+', color='white', s=90, alpha=0.5)
```

```python
# STEP 0
# Generate a 'real' PSF like we'd see on the detector, using the OPD map
input_e_field = np.real(pad_pupil) * np.exp(1.0j * 2 * np.pi * pad_input_opd_map)
initial_psf = np.abs(np.fft.fftshift(np.fft.fft2(input_e_field)))
no_opd_psf = np.abs(np.fft.fftshift(np.fft.fft2(pad_pupil)))

plt.figure(figsize=(30, 7.5))
plt.subplot(1, 3, 1)
plt.title("OPD Map (MIRI_OPDisim6TiltCorrected)")
plt.imshow(np.ma.masked_array(pad_input_opd_map, mask=(pad_pupil <= 0)), cmap='rainbow')
cbar = plt.colorbar()
cbar.set_label('Optical Path Difference')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
plt.xlim(512,1536)
plt.ylim(512,1536)

plt.subplot(1, 3, 2)
plt.title("Simulated PSF Intensity (with OPD)")
plt.imshow(np.abs(initial_psf) / np.max(np.abs(initial_psf)), norm=LogNorm())
cbar = plt.colorbar()
zoom_it(14, initial_psf)
cbar.set_label('Relative Intensity')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')

plt.subplot(1, 3, 3)
plt.title("Simulated PSF Intensity (without OPD)")
plt.imshow(no_opd_psf / np.max(no_opd_psf), norm=LogNorm())
cbar = plt.colorbar()
zoom_it(14, initial_psf)
cbar.set_label('Relative Intensity')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

Now we have a point-spread function (`initial_psf`), simulating one that we would see on a detector. The phase information is lost in the conversion to a real-valued image like we would see on the detector, so we're left to make an initial guess at the phase in the intensity image.

We have the shape of the pupil, but not the phase of the light coming in at each point. Starting with our aperture function, we get the magnitude and phase of the electric field at each point `[j, k]` by computing $$\vec{E}_{jk} = \mathrm{Aperture}_{jk} \ e^{i \phi_{jk}}$$ where $\phi$ is a phase from $-\pi$ to $\pi$. (n.b. The Gerchberg-Saxton algorithm works best to recover small phase offsets, so we initialize a random "guess" array of phases from $-\frac{\pi}{4}$ to $\frac{\pi}{4}$.)

Then, the forward FFT of this electric field array will show us the magnitude and phase at each point on the image plane.

```python
N_ITERATIONS = 21
# Generate an initial guess for the E field in the pupil plane using phases from [-π, π]
phase_guess = np.random.uniform(-0.25*np.pi, 0.25*np.pi, size=initial_psf.shape)
e_for_pupil_corrected_transmittance = np.real(pad_pupil) * np.exp(1.0j * phase_guess)

for i in range(N_ITERATIONS):
    print i,
    plot_flag = False # to see plots: True if i % 2 == 0 else False
    # STEP 1
    # FFT the pupil E field to get a rough E field for the PSF (image plane)
    e_for_image_plane = np.fft.fftshift(np.fft.fft2(e_for_pupil_corrected_transmittance))
    
    if plot_flag:
        # Compare rough to 'real' psf
        plt.figure(figsize=(30,7.5))

        plt.subplot(1, 4, 1)
        plt.title(u'Iteration {0}: Rough PSF from guessed phase'.format(i))
        img1 = plt.imshow(np.abs(e_for_image_plane), norm=LogNorm())
        img1.set_clim(1e-3, 5e5)
        plt.colorbar()
        #zoom_it(10, e_for_image_plane)
        plt.xlabel('X pixel')
        plt.ylabel('Y pixel')

    # STEP 2
    # Normalize the complex values in the image to magnitude 1
    image_plane_normalized = e_for_image_plane / np.abs(e_for_image_plane)

    # STEP 3
    # Multiply the complex values by the intensity pattern of our 'real' PSF to get a 'real' image 
    # plane E field with phase information
    image_plane_corrected_intensity = np.abs(initial_psf) * image_plane_normalized
    
    if plot_flag:
        plt.subplot(1, 4, 2)
        plt.title("PSF with corrected intensity pattern\nand preserved phase")
        img2 = plt.imshow(np.abs(initial_psf), norm=LogNorm())
        img2.set_clim(1e-3, 5e5)
        plt.colorbar()

    # STEP 4
    # IFFT the image plane E field to get a rough pupil E field
    e_for_pupil = np.fft.ifft2(np.fft.ifftshift(image_plane_corrected_intensity))
    
    if plot_flag:
        plt.subplot(1, 4, 3)
        plt.title("Rough Pupil from IFFT")
        img2 = plt.imshow(np.abs(e_for_pupil))
        #img2.set_clim(1e-3, 5e5)
        plt.colorbar()

    # STEP 5
    # Normalize the complex values in the pupil E field to magnitude 1
    e_for_pupil_normalized = e_for_pupil / np.abs(e_for_pupil)

    # STEP 6
    # Multiply the complex values by the real pupil (pad_pupil) to get a 'real' pupil
    # plane E field with phase information
    e_for_pupil_corrected_transmittance = np.real(pad_pupil) * e_for_pupil_normalized
    
    if plot_flag:
        plt.subplot(1, 4, 4)
        plt.title("Corrected Pupil with phase from IFFT")
        img2 = plt.imshow(np.abs(e_for_pupil_corrected_transmittance))
        #img2.set_clim(1e-3, 5e5)
        plt.colorbar()
```

```python
# Final values

plt.figure(figsize=(20,15))

plt.subplot(2, 2, 1)
plt.title(u'Rough PSF from guessed phase\n(N = {0} iterations)'.format(N_ITERATIONS))
img1 = plt.imshow(np.abs(e_for_image_plane), norm=LogNorm())
img1.set_clim(1e-3, 5e5)
plt.colorbar()
#zoom_it(10, e_for_image_plane)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')

plt.subplot(2, 2, 2)
plt.title(u'Rough Pupil from IFFT\n(N = {0} iterations)'.format(N_ITERATIONS))
img2 = plt.imshow(np.abs(e_for_pupil))
plt.colorbar()

plt.subplot(2, 2, 3)
plt.title(u'Difference in relative intensity\n(Guessed PSF - Reference PSF)')
norm_guessed = np.abs(e_for_image_plane) / np.max(np.abs(e_for_image_plane))
norm_reference = np.abs(initial_psf) / np.max(np.abs(initial_psf))
img3 = plt.imshow(norm_guessed - norm_reference, cmap='jet')
img3.set_clim(-1, 1)
plt.text(50, 200, "Mean of difference = {0:1.5f}".format(np.average(norm_guessed - norm_reference)))
plt.text(50, 150, "StdDev of difference = {0:1.3f}".format(np.std(norm_guessed - norm_reference)))
plt.colorbar()
plt.xlabel('X pixel')
plt.ylabel('Y pixel')

plt.subplot(2, 2, 4)
plt.title(u'Difference in pupil transmittance\n(Guessed pupil - Reference pupil)')
norm_pupil_guess = np.abs(e_for_pupil) / np.max(np.abs(e_for_pupil))
img4 = plt.imshow(norm_pupil_guess - np.real(pad_pupil), cmap='jet')
img4.set_clim(-1, 1)
plt.text(50, 200, "Mean of difference = {0:1.5f}".format(np.average(norm_pupil_guess - np.real(pad_pupil))))
plt.text(50, 150, "StdDev of difference = {0:1.3f}".format(np.std(norm_pupil_guess - np.real(pad_pupil))))
plt.colorbar()
```

```python
recovered_phase = np.real(np.log(e_for_pupil_corrected_transmittance) / 1.0j)
recovered_phase = np.ma.masked_array(recovered_phase, mask=(pad_pupil <= 0))

plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
img1 = plt.imshow(np.ma.masked_array(2 * np.pi * pad_input_opd_map, mask=(pad_pupil <= 0)), cmap='jet')
print img1.get_clim()
cbar = plt.colorbar()
cbar.set_label('Optical Path Difference')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
plt.xlim(512,1536)
plt.ylim(512,1536)

plt.subplot(1, 2, 2)
img2 = plt.imshow(recovered_phase, cmap='jet')
img2.set_clim(img1.get_clim())
plt.xlim(512, 1536)
plt.ylim(512, 1536)
cbar = plt.colorbar()
cbar.set_label('Phase Difference')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')


```

```python
# TRYING WITH NP.ANGLE INSTEAD OF NORMALIZING

# Generate an initial guess for the E field in the pupil plane using phases from [-π, π]
phase_guess = np.random.uniform(-0.25*np.pi, 0.25*np.pi, size=initial_psf.shape)
e_for_pupil_corrected_transmittance = np.real(pad_pupil) * np.exp(1.0j * phase_guess)
e_for_pupil_corrected_transmittance2 = np.real(pad_pupil) * np.exp(1.0j * phase_guess)
```

```python
N_ITERATIONS = 21

for i in range(N_ITERATIONS):
    plot_flag = False #True if i % 5 == 0 else False
    # STEP 1
    # FFT the pupil E field to get a rough E field for the PSF (image plane)
    e_for_image_plane = np.fft.fftshift(np.fft.fft2(e_for_pupil_corrected_transmittance))
    e_for_image_plane2 = np.fft.fftshift(np.fft.fft2(e_for_pupil_corrected_transmittance2))
    
    if plot_flag:
        # Compare rough to 'real' psf
        plt.figure(figsize=(30,7.5))

        plt.subplot(1, 4, 1)
        plt.title(u'Iteration {0}: Rough PSF from guessed phase'.format(i))
        img1 = plt.imshow(np.abs(e_for_image_plane), norm=LogNorm())
        img1.set_clim(1e-3, 5e5)
        plt.colorbar()
        #zoom_it(10, e_for_image_plane)
        plt.xlabel('X pixel')
        plt.ylabel('Y pixel')

    # STEP 2
    # Normalize the complex values in the image to magnitude 1
    image_plane_normalized = e_for_image_plane / np.abs(e_for_image_plane)
    image_phase_angles = np.angle(e_for_image_plane)

    # STEP 3
    # Multiply the complex values by the intensity pattern of our 'real' PSF to get a 'real' image 
    # plane E field with phase information
    image_plane_corrected_intensity = np.abs(initial_psf) * image_plane_normalized
    image_plane_corrected_intensity2 = np.abs(initial_psf) * np.exp(1.0j * image_phase_angles)
    
    if plot_flag:
        plt.subplot(1, 4, 2)
        plt.title("PSF with corrected intensity pattern\nand preserved phase")
        img2 = plt.imshow(np.abs(initial_psf), norm=LogNorm())
        img2.set_clim(1e-3, 5e5)
        plt.colorbar()

    # STEP 4
    # IFFT the image plane E field to get a rough pupil E field
    e_for_pupil = np.fft.ifft2(np.fft.ifftshift(image_plane_corrected_intensity))
    e_for_pupil2 = np.fft.ifft2(np.fft.ifftshift(image_plane_corrected_intensity2))
    
    if plot_flag:
        plt.subplot(1, 4, 3)
        plt.title("Rough Pupil from IFFT")
        img2 = plt.imshow(np.abs(e_for_pupil))
        #img2.set_clim(1e-3, 5e5)
        plt.colorbar()

    # STEP 5
    # Normalize the complex values in the pupil E field to magnitude 1
    e_for_pupil_normalized = e_for_pupil / np.abs(e_for_pupil)
    pupil_phase_angles = np.angle(e_for_pupil2)

    # STEP 6
    # Multiply the complex values by the real pupil (pad_pupil) to get a 'real' pupil
    # plane E field with phase information
    e_for_pupil_corrected_transmittance = np.real(pad_pupil) * e_for_pupil_normalized
    e_for_pupil_corrected_transmittance2 = np.real(pad_pupil) * np.exp(1.0j * pupil_phase_angles)
    
    if plot_flag:
        plt.subplot(1, 4, 4)
        plt.title("Corrected Pupil with phase from IFFT")
        img2 = plt.imshow(np.abs(e_for_pupil_corrected_transmittance))
        #img2.set_clim(1e-3, 5e5)
        plt.colorbar()
```

```python
recovered_phase = np.real(np.log(e_for_pupil_corrected_transmittance) / 1.0j)
recovered_phase = np.ma.masked_array(recovered_phase, mask=(pad_pupil <= 0))

recovered_phase2 = np.real(np.log(e_for_pupil_corrected_transmittance2) / 1.0j)
recovered_phase2 = np.ma.masked_array(recovered_phase2, mask=(pad_pupil <= 0))

plt.figure(figsize=(30, 10))
plt.subplot(1, 3, 1)
plt.title("Reference Phase from OPD")
img1 = plt.imshow(2 * np.pi * np.ma.masked_array(pad_input_opd_map, mask=(pad_pupil <= 0)), cmap='rainbow')
img1.set_clim(-0.25*np.pi,0.25*np.pi)
cbar = plt.colorbar()
cbar.set_label('Phase Difference (-pi/4, pi/4)')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
plt.xlim(512,1536)
plt.ylim(512,1536)

plt.subplot(1, 3, 2)
plt.title("Recovered Phase (normalize and multiply method)")
img2 = plt.imshow(recovered_phase, cmap='rainbow')
img2.set_clim(-0.25*np.pi, 0.25*np.pi)
plt.xlim(512, 1536)
plt.ylim(512, 1536)
cbar = plt.colorbar()
cbar.set_label('Phase Difference (-pi/4, pi/4)')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')

plt.subplot(1, 3, 3)
plt.title("Recovered Phase (np.angle method)")
img2 = plt.imshow(recovered_phase2, cmap='rainbow')
img2.set_clim(-0.25*np.pi, 0.25*np.pi)
plt.xlim(512, 1536)
plt.ylim(512, 1536)
cbar = plt.colorbar()
cbar.set_label('Phase Difference (-pi/4, pi/4)')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')


```
