> # Starting from the JWST pupil, calculate its Fourier transform and plot the corresponding PSF.

OPD files and a FITS pupil image are available in `/grp/jwst/wit/miri/lajoie`.

```python
pupil = fits.getdata('./pupil.fits')
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(pupil)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

```python
def zoom_it(frac, imgdata):
    centerx, centery = imgdata.shape[:2]
    centerx /= 2
    centery /= 2
    frac = 1.0 / frac
    plt.xlim(centerx - (frac * float(centerx)), centerx + (frac * float(centerx)))
    plt.ylim(centery - (frac * float(centery)), centery + (frac * float(centery)))
    plt.scatter((centerx,), (centery,), marker='+', color='white', s=90, alpha=0.5)
```

```python
psf = np.fft.fftshift(np.fft.fft2(pupil))
```

```python
plt.figure(figsize=(9.25,7.25))
img = plt.imshow(np.abs(psf) / np.max(np.abs(psf)), norm=LogNorm())
cbar = plt.colorbar()
cbar.set_label('Relative Intensity')
zoom_it(12, psf)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

> # Why can't we simply use `numpy.fft.fft2`?

We aren't sampling the PSF very finely in the region we are most interested in: that of highest intensity in the core. If we have a pupil that spans the input array from edge to edge, we'll have a very small PSF. Instead, we must pad the pupil array in order to get more detail in our PSF output.

```python
# if we pad by 1024 + 512, we preserve the useful property of dimensions being a power of two for the FFT (total side dimension: 4096)
pad_pupil = np.pad(pupil, 1024 + 512, mode='constant', constant_values=(0.,)).astype(np.float64, casting='safe')
print pad_pupil.shape, pad_pupil.dtype
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.real(pad_pupil))
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

```python
pad_psf = np.fft.fftshift(np.fft.fft2(pad_pupil))
print pad_psf.shape, pad_psf.dtype
```

```python
plt.figure(figsize=(25,10))

plt.subplot(1, 2, 1)
plt.title(u'PSF (pupil occupied 1024×1024 of 1024×1024 input array)')
img = plt.imshow(np.abs(psf) / np.max(np.abs(psf)), cmap='hot')
plt.colorbar()
zoom_it(16, psf)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')

plt.subplot(1, 2, 2)
plt.title(u'PSF (pupil occupied 1024×1024 of 4096×4096 input array)')
img = plt.imshow(np.abs(pad_psf) / np.max(np.abs(pad_psf)), cmap='hot')
plt.colorbar()
zoom_it(16, pad_psf)
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
```

By padding each side of the pupil array by 1536, we preserve the power-of-two dimensions (saving us computation time) and we improve the sampling of the core of the PSF.
